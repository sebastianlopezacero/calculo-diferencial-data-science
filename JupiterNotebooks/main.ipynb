# L√≠mites

## ¬øQu√© es un l√≠mite?

Un l√≠mite es un an√°lisis de tendencia respecto a un punto. Es decir, hacia donde va el valor de nuestra funci√≥n a medida que ‚Äúx‚Äù se acerca al valor ‚Äúa‚Äù, pero sin tocar nunca ese valor ‚Äúa‚Äù. Por eso se llama l√≠mite, porque ‚Äútenemos prohibido tocar‚Äù ese valor ‚Äúa‚Äù.

$$ \lim_{{x \to a}} f(x) = L $$

Una analog√≠a para entender el l√≠mite es la siguiente: sup√≥n que tienes una fogata encendida frente a t√≠, y quieres comprobar que el fuego que est√°s viendo existe, que tus ojos no te enga√±an. Para ello, acercas las manos al fuego, cada vez m√°s despacio, a ver si las sientes m√°s calientes, pero nunca llegas a estar directamente en el fuego porque te puedes quemar.

Para resolver un l√≠mite veamos el concepto de l√≠mites laterales. Un l√≠mite lateral es cuando nos acercamos a un valor por un lado, ya sea por la izquierda o la derecha. Los l√≠mites laterales se denotan de la siguiente manera.

$$ \lim_{{x \to a^+}} f(x) = L $$

Esto es un l√≠mite lateral a la izquierda. Un l√≠mite lateral a la derecha se denota del mismo modo, excepto que elevamos la constante ‚Äúa‚Äù al signo positivo.

$$ \lim_{{x \to a^-}} f(x) = L $$


Los l√≠mites laterales son importantes, ya que para que el l√≠mite en general exista, los dos l√≠mites laterales deben ser iguales.

# Derivada en ciencia de datos


## Definici√≥n de la derivada

El problema de la derivada nace de tratar de encontrar la recta tangente a un punto en una curva. La primera soluci√≥n (por parte de Isaac Newton) fu√© tomar una recta secante, es decir una recta que corta a la curva en dos puntos.


De la imagen anterior nos damos cuenta que estos dos puntos est√°n dados por

$$ {P_{1}} = (x,f(x)) $$

y

$$ {P_{2}} = ((x+h),f(x+h)) $$

donde ‚Äúh‚Äù es la distancia horizontal entre dichos puntos. Mediante estos dos puntos, podemos calcular la pendiente de la recta secante con la f√≥rmula

$$ m = \frac{y_{2}-y_{1}}{x_{2}-x_{1}} $$

donde m es la pendiente.

Sin embargo, esto nos da la pendiente de la recta secante. Queremos encontrar la de la recta tangente. Para ello, debemos recortar la distancia ‚Äúh‚Äù, hasta que sea muy cercana a cero.

Entonces tomamos el l√≠mite, y esto nos da la pendiente de la recta tangente en un punto x de la curva. A este l√≠mite es lo que llamamos derivada.

$$ \lim_{{h \to 0}} = \frac{f(x+h)-f(x)}{h}  $$

La derivada tambi√©n se puede ver en t√©rminos de incrementos. El numerador ser√≠a el incremento $\delta{y}$ entre las funciones o tambi√©n el valor de y, mientras que el denominador ser√≠a el incremento $\delta{x}$ entre los valores de x.



## La derivada como raz√≥n de cambio

Sup√≥n que llevas un tiempo estudiando la relaci√≥n entre los a√±os de experiencia como desarrollador de software (developer) y el salario de este trabajo. Llegas a la conclusi√≥n de que esta relaci√≥n es cuadr√°tica, y que el gr√°fico se ve como una par√°bola. Llamemos ‚Äús‚Äù al salario y ‚Äúa‚Äù a la cantidad de a√±os de experiencia. La relaci√≥n entre ambas variables ser√≠a $s(a)=a^2$ 

Si quisieras saber que tanto sube el salario en funci√≥n de los a√±os, podemos calcular la derivada. Esta nos va a decir que tan r√°pido cambia ‚Äús‚Äù respecto a ‚Äúa‚Äù. La derivada de una funci√≥n cuadr√°tica $f(x)=x^2$  es simplemente $f¬¥(x)=x^2$ , por lo que en este caso nos queda $s¬¥(a)=a^2$

 En general, la derivada de una funci√≥n $f(x)=x^n$ es $f¬¥(x)=nx^{n-1}$. Aqu√≠ puedes ver una tabla de drivadas comunes.

## Implementaci√≥n de la derivada discreta

```python
# Funci√≥n inicial f(x)
def f(x):
    return x**2

x = np.linspace(-10,10,num=1000)
y = f(x)

# Calculando la deriva de la funci√≥n
def df(f):
    h = 0.0000000000001
    return (f(x+h)-f(x))/h
# Graficando la funci√≥n
plt.plot(x,y, "b")
plt.plot(x,df(f),"r--", label="Derivada 1")
plt.legend()
plt.grid()
```

# Introducci√≥n a m√°ximos y m√≠nimos


## ¬øQu√© es un m√°ximo y un m√≠nimo?

En matem√°ticas, un m√°ximo y un m√≠nimo son valores extremos de una funci√≥n que representan los puntos m√°s altos y m√°s bajos, respectivamente. Estos extremos son importantes en el an√°lisis de funciones, ya que brindan informaci√≥n sobre los puntos cr√≠ticos y las caracter√≠sticas de la funci√≥n. A continuaci√≥n, se describen con m√°s detalle cada uno de ellos:

- *M√°ximo:*
Un m√°ximo de una funci√≥n es el valor m√°s alto que la funci√≥n alcanza en un determinado intervalo o en todo su dominio. Formalmente, sea f una funci√≥n definida en un conjunto D, se dice que f tiene un m√°ximo en el punto c si $f(c)$ es mayor o igual que $f(x)$ para todo x en D cercano a c. Un m√°ximo absoluto es aquel que es el valor m√°s alto de la funci√≥n en todo su dominio, mientras que un m√°ximo relativo es aquel que es el valor m√°s alto en un intervalo espec√≠fico.
- *M√≠nimo:*
Un m√≠nimo de una funci√≥n es el valor m√°s bajo que la funci√≥n alcanza en un determinado intervalo o en todo su dominio. Formalmente, sea f una funci√≥n definida en un conjunto D, se dice que f tiene un m√≠nimo en el punto c si $f(c)$ es menor o igual que $f(x)$ para todo x en D cercano a c. Un m√≠nimo absoluto es aquel que es el valor m√°s bajo de la funci√≥n en todo su dominio, mientras que un m√≠nimo relativo es aquel que es el valor m√°s bajo en un intervalo espec√≠fico.

### Funci√≥n escal√≥n o de Heaviside

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$f¬¥(x)=0$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N=1000


y = H(x)

plt.plot(x, H(x), 'b')
plt.plot(x,df(H), 'r')
plt.grid()
```

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```
